A vehicle is counted when:

- no very similar detection has been counted in the last few frames near the same x-position.
</details>

<details>
<summary>👀 Need help? Solution Outline</summary>

- Maintain a list of **recently counted positions** (e.g. x-coordinates of vehicle centers).
- For every frame:
  
  1. **Update the memory of previously counted vehicles**
     - Reduce a small time-to-live counter for each stored position.
     - Remove entries whose lifetime has expired.

  2. **Process the detections**
     - For each detected bounding box, compute its **centroid**.

  3. **Check if the vehicle is near the counting line**
     - Verify that the vertical distance between the centroid and the line is smaller than a small threshold (a "band" around the line).

  4. **Avoid double counting**
     - Check whether a recently counted vehicle already exists **near the same horizontal position**.
     - If the centroid is too close to an already counted one, ignore it.

  5. **Count the vehicle**
     - If the detection is close to the line **and** not already counted nearby:
       - increase the vehicle counter
       - store the centroid x-position in the list of active counted vehicles.

This simple strategy prevents counting the same vehicle multiple times while it remains close to the counting line.
